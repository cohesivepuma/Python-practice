```python
#python 中的数据类型
x = 10 #int
y = 1.1 #float
z = 8 + 8j #complex
s = "hello world" #str
arr = [114,514] #list 有序数组
arr_un = (114,514,1919) #tuple 不可被修改的有序数组
arr_only = {1,2,3,4} #set 不可重复的无序数组
g = {'name':'tom','age':18} #dictionary 一对键值对
right = True #布尔函数
wrong = False
none = None #空值
print(type(x))
print(type(y))
print(type(z))
print(type(s))
print(type(arr))
print(type(arr_un))
print(type(arr_only))
print(type(g))
print(type(right))
print(type(wrong))
print(type(none))
```
